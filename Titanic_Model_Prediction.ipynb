```python
#Import Required Library For Titanic Dataset

import numpy as np
import pandas as pd
```

```python
# Import Titanic Dataset 

dataset = pd.read_csv('titanic_dataset_2.csv')
```

```python
#Visualize Head Of Imported Dataset

dataset.head(5)
```

```python
#Visualize Tail Of Imported Dataset

dataset.tail(5)
```

```python
#Cheack Dataset Clean Or Not (Note :- If Not Then Clean It)

dataset.isnull()
```

```python
#Cheack Dataset Clean Or Not In Sum Formate (Note :- If Not Then Clean It)

dataset.isnull().sum()
```

```python
#Check Discription Of Dataset

dataset.describe()
```

```python
#Evaluate Column Value For Better Understanding

dataset.nunique()
```

```python
#Set Independent Value From Imported Dataset

x = dataset.iloc[ : ,1:8]
x
```

```python
#Set Dependent Value From Imported Dataset

y = dataset.iloc[ : ,0:1]
y
```

```python
#If Data Will Be In Formate Then You Have To Be Converted In Number Formate


#from sklearn.preprocessing import LabelEncoder
#label_encoder = LabelEncoder()
#y = label_encoder.fit_transform(y)
```

```python
#Split Your Dataset In (Train , Test) Formate 

from sklearn.model_selection import train_test_split
x_train , x_test , y_train ,y_test = train_test_split(x , y , test_size = 0.469)
```

```python
#Applied Model A/Q To Visibility

from sklearn.ensemble import RandomForestClassifier
log_model = RandomForestClassifier()
log_model.fit(x_train , y_train)
```

```python
#Predict Your Model

y_pred = log_model.predict(x_test)

y_pred
```

```python
#Vsualized iNDataFrame Formate

reformed = pd.DataFrame(np.c_[x_test , y_test , y_pred])
reformed
```

```python
#Check The Shape

y_pred.shape
```

```python
#Check The Accuracy

from sklearn.metrics import confusion_matrix 
confusion_matrix(y_test , y_pred)
```
